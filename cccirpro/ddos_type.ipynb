```python
import pandas as pd
```

```python
df = pd.read_csv('WSNBFSFdataset.csv')
```

```python
df.head()
```

```python
X = df[['Time',	'S_Node',	'Node_id',	'Rest_Energy',	'Trace_Level',	'Mac_Type_Pckt',	'Source_IP_Port',	'Des_IP_Port', 'Packet_Size',	'TTL',	'Hop_Count',	'Broadcast_ID',	'Dest_Node_Num',	'Dest_Seq_Num',	'Src_Node_ID',	'Src_Seq_Num']]
y = df[['Class']]
```

```python
from sklearn import preprocessing
lab_enc = preprocessing.LabelEncoder()
y = lab_enc.fit_transform(y)
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, y_train, x_test, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

model = DecisionTreeClassifier(max_leaf_nodes=3)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
acc = accuracy_score(y_pred, y_test)
print(acc)
```
